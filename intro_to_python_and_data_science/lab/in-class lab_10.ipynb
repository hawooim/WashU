# Plotting Practice


In this practice, we will analyze various statistics from MBA applicants across 22 different industries.  The data is contained in the file "MBA_Stat.csv". 


First, read in the data and use the head method to double check that everything has been read in correctly.  Make sure to add you "%matplotlib inline" statement at the top of the code block as well importing pandas and matplotlib.pyplot.

```python
%matplotlib inline

#Import here
import pandas as pd
import matplotlib.pyplot as plt

#Set you style if you want
plt.style.use('ggplot')

#Write your code here
df = pd.read_csv("./data/MBA_Stats.csv", index_col = 0)
df.head()
```

We will begin by creating some basic single plots.  First create a scatterplot of Avg. Age vs. Avg. GMAT Score.

```python
#Create figure and axis
fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (5, 5))

#Create a plot
df.plot(kind = 'scatter', x = 'Avg_Age', y = 'Avg_GMAT', ax = ax)
fig.suptitle("Average Age vs. Average GMAT Score", size = 13)
ax.set(xlabel = "Average Age", ylabel = "Average GMAT Score")
```

Next, we would like to create the following two histograms, which give us how the average GMAT and average previous work experience is distributed across industries.

 <img src="distributions.png">

First create the 1x2 figure.  Use figsize = (10,5). Add the figure title.

```python
#Create figure and axis
fig, (ax0, ax1) = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 5))
fig.suptitle("Distribution of GMAT and Work Experience", size = 15)
```

Next create th first histogram on ax0.  Vary the bin size so the histogram provides as much information as possible. Also, add the appropriate x-axis title.

```python
#Create the histogram
df.plot(kind = 'hist', y ='Avg_GMAT', color = 'r', bins = 10, ax = ax0)

#Add x-axis title
ax0.set(xlabel = 'Avg. GMAT')
ax0.get_legend().remove()
fig
```

Now add in the vertical dashed average line with the corresponding annotation.

```python
#Add vertical average line
ax0.axvline(df['Avg_GMAT'].mean(), color = 'b', linestyle = '--', linewidth = 1)
ax0.annotate("Average = %.2f" %df['Avg_GMAT'].mean(), xy = (df['Avg_GMAT'].mean()+2, 3.5))
fig
```

Now create the second histogram for work experience

```python
#Create the histogram
df.plot(kind = 'hist', y = 'Avg_Job_Months', color = 'g', bins = 10, ax = ax1)

#Add x-axis title
ax1.set(xlabel = "Avg. Months of Work Experience")
ax1.get_legend().remove()

#Add vertical average line
ax1.axvline(df['Avg_Job_Months'].mean(), color = 'b', linestyle = '--', linewidth = 1)
ax1.annotate("Average = %.2f" %df['Avg_Job_Months'].mean(), xy = (df['Avg_Job_Months'].mean()+1, 3.5))

fig
```

Create a 1X3 figure with a shared y axis and three axes. Set figsize = (10,10). Add the figure title to the figure.

```python
#Write your code here
fig, (ax0, ax1, ax2) = plt.subplots(nrows = 1, ncols = 3, sharey = True, figsize = (10, 10))
fig.suptitle("MBA Applicants Statistics by Industry", size = 15)
```

Create the plot on ax0, you should set kind = "barh". Add the appropriate titles and axes labels.

```python
#Add the plot
df.plot(kind = 'barh', x = 'Industry', y = 'Avg_GMAT', color = 'r', ax = ax0)

#Update axes features.  Make the x axis go from 600 to 690.
ax0.set(title = 'GMAT', xlabel = 'Avg. GMAT Score', xlim = (600, 690))
fig
```

Create the plot of ax1.

```python
#Add the plot
df.plot(kind = 'barh', x = 'Industry', y = 'Avg_Age', color = 'g', ax = ax1)

#Update axes features.  Make the x axis go from  25 to 32.
ax1.set(title = 'Age', xlabel = 'Avg. Age', xlim = (25, 32))
fig
```

Create the plot of ax2.

```python
#Add the plot
df.plot(kind = 'barh', x = 'Industry', y = 'Avg_Job_Months', color = 'b', ax = ax2)

#Update axes features.  Make the x axis go from  45 to 75.
ax2.set(title = 'Work Experience', xlabel = 'Avg. Months Work Experience', xlim = (45, 75))
fig
```

```python
# Add averages (just for fun)
ax0.axvline(df['Avg_GMAT'].mean(), color = 'black', linestyle = '--', linewidth = 1)
ax0.annotate("Avg. = %.1f" %df['Avg_GMAT'].mean(), xy = (df['Avg_GMAT'].mean()+2, 11))

ax1.axvline(df['Avg_Age'].mean(), color = 'black', linestyle = '--', linewidth = 1)
ax1.annotate("Avg. = %.1f" %df['Avg_Age'].mean(), xy = (df['Avg_Age'].mean()+0.15, 11))

ax2.axvline(df['Avg_Job_Months'].mean(), color = 'black', linestyle = '--', linewidth = 1)
ax2.annotate("Avg. = %.1f" %df['Avg_Job_Months'].mean(), xy = (df['Avg_Job_Months'].mean()+0.8, 11))

fig
```

```python
# Remove legends because they block the graphs

ax0.get_legend().remove()
ax1.get_legend().remove()
ax2.get_legend().remove()

fig
```

```python
# Plot vertical bar graphs
fig, (ax0, ax1, ax2) = plt.subplots(nrows = 3, ncols = 1, sharex = True, figsize = (10, 10))
fig.suptitle("MBA Applicants Statistics by Industry", size = 15)
```

```python
df.plot(kind = 'bar', x = 'Industry', y = 'Avg_GMAT', color = 'r', ax = ax0)
ax0.set(title = 'GMAT', ylabel = 'Avg. GMAT Score', ylim = (600, 690))

df.plot(kind = 'bar', x = 'Industry', y = 'Avg_Age', color = 'g', ax = ax1)
ax1.set(title = 'Age', ylabel = 'Avg. Age', ylim = (25, 32))

df.plot(kind = 'bar', x = 'Industry', y = 'Avg_Job_Months', color = 'b', ax = ax2)
ax2.set(title = 'Work Experience', ylabel = 'Avg. Months Work Experience', ylim = (45, 75))

ax0.axhline(df['Avg_GMAT'].mean(), color = 'black', linestyle = '--', linewidth = 1)
ax0.annotate("Avg. = %.1f" %df['Avg_GMAT'].mean(), xy = (6, df['Avg_GMAT'].mean()))

ax1.axhline(df['Avg_Age'].mean(), color = 'black', linestyle = '--', linewidth = 1)
ax1.annotate("Avg. = %.1f" %df['Avg_Age'].mean(), xy = (8, df['Avg_Age'].mean()))

ax2.axhline(df['Avg_Job_Months'].mean(), color = 'black', linestyle = '--', linewidth = 1)
ax2.annotate("Avg. = %.1f" %df['Avg_Job_Months'].mean(), xy = (9, df['Avg_Job_Months'].mean()))

fig
```
